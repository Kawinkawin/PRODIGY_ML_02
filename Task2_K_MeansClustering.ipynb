```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
df=pd.read_csv("/content/Mall_Customers.csv")
```

```python
df.head(10)
```

```python
cp_df=df.copy()
df.shape
df.isnull().sum()
```

```python
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.title('Relationship between Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.show()
```

```python
X=cp_df.iloc[:,[3,4]].values
```

```python
# CALCULATING Within-Cluster Sum of Square
wcss=[]
for i in range(1,11):
       kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
       kmeans.fit(X)
       wcss.append(kmeans.inertia_)
```

```python
#elbow movement
plt.gca().set_facecolor('black')
plt.plot(range(1,11),wcss,linewidth=3)
plt.title('The Elbow Movement')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
silhouette_scores = []

for n_clusters in range(2, 11):  # Start from 2 clusters
    # Initialize and fit KMeans model
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(X)

    # Calculate silhouette score
    silhouette_avg = silhouette_score(X, labels)
    silhouette_scores.append(silhouette_avg)
    print("Number of clusters:", n_clusters, "| Silhouette score:", silhouette_avg)
```

```python
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
Y=kmeans.fit_predict(X)
print(Y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green',label='Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red',label='Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow',label='Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='blue',label='Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='black',label='Cluster 5')
```
